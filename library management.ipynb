```python
import sqlite3

    
def connect_database():
    try:
        
        con=sqlite3.connect("library.db")
        cur=con.cursor()
        cur.execute('''
                create table if not exists books
                ( book_id integer primary key,
                  author text not null,
                  title text not null,
                  quantity integer not null)
                    ''')
        return con,cur
    except sqlite3.Error as e:
        print("Error creating connection:", e)
        return None, None


def close_connection(con, cur):
    if cur:
        cur.close()
    if con:
        con.close()
def add_book():
    con,cur=connect_database()
    try:
        
        author=input("enter the author of the book:")
        title=input("enter the title of the book:")
        quantity=int(input("enter the quantity:"))
        cur.execute('''insert into books (author,title,quantity) values(?,?,?)''',(author,title,quantity))
        con.commit()
        print("book added successfully")
    except sqlite3.Error as error:
        print("error:",error)
    finally:
        close_connection(con, cur)
def update_book():
    con,cur=connect_database()
    try:
        book_id=input("enter the id of the book: ")
        new_author=input("enter the new author:")
        new_title=input("enter the new title:")
        new_quantity=int(input("enter the new quantity:"))
        cur.execute('''update books set author=?,
                        set title=?,
                        set quantity=?
                        where book_id=book_id''',(new_author,new_title,new_quantity))
    except sqlite3.Error as error:
        print("error",error)
    finally:
        close_connection(con, cur)
def search_bookbyid():
    con,cur=connect_database()
    try:
        book_id=input("enter the book id to search:")
        cur.execute('''select * from books where book_id=?''',(book_id,))
        book=cur.fetchone()
        if book:
            print("bookid:",book[0])
            print("book author:",book[1])
            print("book title:",book[2])
            print("book quantity:",book[3])
        else:
            print("book not found")
    except sqlite3.Error as error:
        print("error",error)
    finally:
        close_connection(con, cur)

def search_bookbyauthor():
    con,cur=connect_database()
    try:
        author=input("enter the author name:")
        cur.execute('''select *from books where author like ?''',('%'+author+'%',))
        book1=cur.fetchall()
        if book1:
            for book in book1:
                print("book id:",book[0])
                print("book author:",book[1])
                print("book title:",book[2])
                print("book quantity:",book[3])
        else:
            print("not found")
    except sqlite3.Error as error:
        print("error",error)
    finally:
        close_connection(con, cur)
def delete_book():
    con,cur=connect_database()
    try:
        book_id=int(input("enter the book id to delete book:"))
        cur.execute('''select * from books where book_id=?''',(book_id,))
        book=cur.fetchone()
        if book:
            cur.execute('''delete from books where book_id=?''',(book_id,))
            con.commit()
            print("book deleted successfully")
        else:
            print("book not found")
    except sqlite3.Error as error:
        print("error",error)
    finally:
        close_connection(con, cur)
def display():
    con,cur=connect_database()
    try:
        cur.execute('''select *from books''')
        bookfinal=cur.fetchall()
        if bookfinal:
            
            for bookss in bookfinal:
                print(bookss)
        else:
            print("No books present")
            
    except sqlite3.Error as error:
        print("error",error)
    finally:
        close_connection(con, cur)


while True:
    print("Menu for Library management")
    print("1.add book\n")
    print("2.update book\n")
    print("3.search book by id\n")
    print("4.search book by author\n")
    print("5.delete book\n")
    print("6.display\n")
    choice=int(input("enter the choice: "))
    if choice==1:
        add_book()
    elif choice==2:
        update_book()
    elif choice==3:
        search_bookbyid()
    elif choice==4:
        search_bookbyauthor()
    elif choice==5:
        delete_book()
    elif choice==6:
        display()
    else:
        print("wrong")
        break
```
